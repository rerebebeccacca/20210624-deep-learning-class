```python
import pickle
import keras
from keras.models import load_model
import cv2
from glob import glob

import matplotlib.pyplot as plt
import numpy as np
import pickle
import os
```

```python
folder_path = '/content/drive/MyDrive/勞動部/03-CNN-Applications/Face_recognition/new'
```

```python
paths = glob(os.path.join(folder_path, 'new/*'))
```

```python
len(paths)
```

```python
img = cv2.imread(paths[1])[:,:,::-1]
plt.imshow(img)
```

```python
model = load_model(os.path.join(folder_path, 'facenet.h5'))
```

```python
model.summary()
```

```python
X = np.zeros((len(paths), 160, 160, 3))
for i, path in enumerate(paths):
    img = cv2.imread(path)[:,:,::-1]
    img = cv2.resize(img, (160, 160))
    X[i] = img
```

```python
def preprocess(imgs):
    for i in range(imgs.shape[0]):
        # standardization
        img = imgs[i]
        mean, std = img.mean(), img.std()
        img = (img - mean) / std
        imgs[i] = img
    return imgs
X = preprocess(X)
```

```python
embeddings = model.predict(X)
```

```python
embeddings.shape
```

```python
embeddings[0]
```

```python
names = [path.split('/')[-1].split('.')[0] for path in paths]
```

```python
names[:10]
```

```python
db = {}
db['names'] = names
db['embeddings'] = embeddings
```

```python
with open('/content/drive/My Drive/db.pickle', 'wb') as file:
    pickle.dump(db, file)
```
